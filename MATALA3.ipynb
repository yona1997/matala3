```python
import requests
import json
import numpy as nmp
import pandas as pd
```

```python
def get_lat(destination):
    api_key = "AIzaSyD266X3gh1F7ZM-x-adnaid0DLy3Nb1fFc"
    url="https://maps.googleapis.com/maps/api/geocode/json?address=%s&key=%s" % (destination,api_key)
    try:
        response = requests.get(url)
        if not response.status_code == 200:
            print("HTTP error",response.status_code)
        else:
            try:
                response_data = response.json()
                return response_data['results'][0]['geometry']['location']['lat']
            except:
                print("Response not in valid JSON format")
    except:
        print("Something went wrong with requests.get")

def get_lng(destination):

    api_key = "AIzaSyD266X3gh1F7ZM-x-adnaid0DLy3Nb1fFc"
    url="https://maps.googleapis.com/maps/api/geocode/json?address=%s&key=%s" % (destination,api_key)
    try:
        response = requests.get(url)
        if not response.status_code == 200:
            print("HTTP error",response.status_code)
        else:
            try:
                response_data = response.json()
                return response_data['results'][0]['geometry']['location']['lng']
            except:
                print("Response not in valid JSON format")
    except:
        print("Something went wrong with requests.get")

def get_dist_lng(destination):
    origins = "Tel Aviv, Israel"
    api_key = "AIzaSyD266X3gh1F7ZM-x-adnaid0DLy3Nb1fFc"
    url="https://maps.googleapis.com/maps/api/distancematrix/json?origins=%s&destinations=%s&key=%s" % (origins, destination ,api_key)
    try:
        response = requests.get(url)
        if not response.status_code == 200:
            print("HTTP error",response.status_code)
        else:
            try:
                response_data = response.json()
                distance_value = (response_data['rows'][0]['elements'][0]['distance']['value']) / 1000
                return distance_value
            except:
                print("Response not in valid JSON format")
    except:
        print("Something went wrong with requests.get")

def get_dur_lng(destination):
    origins = "Tel Aviv, Israel"
    api_key = "AIzaSyD266X3gh1F7ZM-x-adnaid0DLy3Nb1fFc"
    url="https://maps.googleapis.com/maps/api/distancematrix/json?origins=%s&destinations=%s&key=%s" % (origins, destination ,api_key)
    try:
        response = requests.get(url)
        if not response.status_code == 200:
            print("HTTP error",response.status_code)
        else:
            try:
                response_data = response.json()
                duration_value = (response_data['rows'][0]['elements'][0]['duration']['value']) / 60
                return round(duration_value,2)
            except:
                print("Response not in valid JSON format")
    except:
        print("Something went wrong with requests.get")
```

```python
with open('dests.txt', 'r') as f:
    for line in f:
        destination = f.readlines()
        get_dist_lng(destination)

df = pd.DataFrame({"Target": destination})
df["Target"] = df["Target"].str.strip()
df["Distance_km"] = df["Target"].apply(get_dist_lng)
df["Duration (min)"] = df["Target"].apply(get_dur_lng)
df["Longitude"] = df["Target"].apply(get_lng)
df["latitude"] = df["Target"].apply(get_lat)
```
